### Aleatory Distribution Formula

The formula for a random variable \( X \) with a probability density function \( f(x) \) is given by:

$$ P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx $$

Where:
- \( P(a \leq X \leq b) \) is the probability that \( X \) falls between \( a \) and \( b \).
- \( f(x) \) is the probability density function of \( X \).
- \( \int_{a}^{b} \) denotes the integral from \( a \) to \( b \).



$$ P(A) = \sum P(\{ (e_1,...,e_N) \})  =  {N}\choose{k} \cdot p^kq^{N-k}$$

$P\left( x \right) = \frac{{e^{ - \lambda } \lambda ^x }}{{x!}}$

$P\left( x = k\right) = \frac{{{ - \lambda }^k * e^-\lambda }}{{k!}}$

${\lambda} > 0,  k = 0,1,2,.... $

$ \large f(x) = \frac1{\sigma \sqrt 2_\pi}e^-\frac{} x-\mu^2{2\sigma^2}$

$ f(x) = \frac1{\sigma \sqrt 2_\pi}e^-\frac{(x-\mu)^2}{2\sigma^2}$

$ f(x) = \frac1{\sigma \sqrt 2_\pi}e^-\frac{(x-\mu)^2}{2\sigma^2}$

```python
import numpy as np
```

```python
random_generator=np.random.default_rng()
```

```python
lambds_value=10
sample_size=10000
random_poisson_data_10=random_generator.poisson(lam=lambds_value,size=sample_size)
```

```python
import seaborn as sns
```

```python
def distribution_plotter(data,label,bins='auto'):
    sns.set(rc={"figure.figsize":(7,5)})
    sns.set_style("white")
    dist=sns.histplot(data,bins=bins,stat='density',kde=False,line_kws={'linewidth':5},thresh=None)
    dist.set_title('Distribucion' + label + ' ', fontsize=16)

distribution_plotter(random_poisson_data_10, "Poisson")
```

```python
lambda_value=100
random_poisson_data_100=random_generator.poisson(lam=lambda_value,size=sample_size)
distribution_plotter(random_poisson_data_100,"Poisson")
```

```python
lambda_value=1000
random_poisson_data_1000=random_generator.poisson(lam=lambda_value,size=sample_size)
distribution_plotter(random_poisson_data_1000,"Poisson")
```

```python
lambda_value=10000
random_poisson_data_10000=random_generator.poisson(lam=lambda_value,size=sample_size)
distribution_plotter(random_poisson_data_10000,"Poisson")
```
